```python
#モジュールをインポート
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#保存したcsvを開く
df_helth = pd.read_csv("helth_data.csv")
df_weather = pd.read_csv("weather_data.csv")
```

```python
#データの大きさを確認
print(df_helth.shape)
print(df_weather.shape)
```

```python
df_i = df_helth[df_helth['name'] == '牛澤拓海']
df_i = df_i.drop('date', axis=1)
df_i = df_i.reset_index(drop=True)
```

```python
#ヘルスデータに合わせて天気データを加工
df_weather = df_weather[df_weather['date'] != '12-21']
df_weather = df_weather[df_weather['date'] != '1-3']
df_weather = df_weather[df_weather['date'] != '1-6']
df_weather = df_weather[df_weather['date'] != '1-10']
df_weather = df_weather[df_weather['date'] != '1-13']
df_weather = df_weather[df_weather['date'] != '1-14']
```

```python
#データの大きさを確認(2回目)
print(df_i.shape)
print(df_weather.shape)
```

```python
#データを連結
df = pd.merge(df_i, df_weather, left_index=True, right_index=True)
df
```

```python
#欠損値の確認
df.isnull().sum()
```

```python
corr_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```
